```python
import pandas as pd
from nltk.corpus import words
from pprint import pprint
import itertools
word_list = words.words()

def get_unique_letters(word):
    word = word.lower()
    letters_list = []
    for letter in word:
        if letter not in letters_list:
            letters_list.append(letter)
    return letters_list
```

```python
print('Total words in english:', len(word_list))

five_letter_words = []
for word in word_list:
    if len(word) == 5:
        five_letter_words.append(word)
        
print('Total 5-letter words in english:', len(five_letter_words))
```

```python
alphabet_list = []
for word in five_letter_words:
    for letter in word:
        letter = letter.lower()
        if letter not in alphabet_list:
            alphabet_list.append(letter.lower())
alphabet_list.sort()
print(alphabet_list)
```

```python
frequency_dict = {}
for letter in alphabet_list:
    frequency_dict[letter] = 0

for word in five_letter_words:
    word = word.lower()
    unique_letters = get_unique_letters(word)
    for letter in unique_letters:
        frequency_dict[letter] += 1
```

```python
df_dict = {
    'letter' : [],
    '5letter_word_count' : []
}
for letter, value in frequency_dict.items():
    df_dict['letter'].append(letter)
    df_dict['5letter_word_count'].append(value)
    
freq_df = pd.DataFrame(df_dict).sort_values(by='5letter_word_count', ascending=False)

freq_df['5letter_word_count'].sum()

display(freq_df)
```

```python
def get_position_probability(target_letter, target_position, word_list):
    target_letter = target_letter.lower()
    counter = 0
    for word in word_list:
        word = word.lower()
        letters = list(word)
        if letters[target_position] == target_letter:
            counter += 1
    return counter

round1_remaining_word_scores = {
    'word' : [],
    'remaining_word_count' : [],
    'remaining_words' : [],
    'lett_1' : [],
    'lett_2' : [],
    'lett_3' : [],
    'lett_4' : [],
    'lett_5' : [],
    
    'pos_1' : [],
    'pos_2' : [],
    'pos_3' : [],
    'pos_4' : [],
    'pos_5' : []
}

letter_lookup = freq_df.set_index('letter').to_dict()['5letter_word_count']
counter = 0
for word in five_letter_words:
    remaining_words = []
    word = word.lower()
    letters = get_unique_letters(word)
    for comp_word in five_letter_words:
        comp_word = comp_word.lower()
        letters_matched = 0
        for comp_letter in letters:
            if comp_letter in comp_word:
                letters_matched += 1
        if letters_matched == 0:
            remaining_words.append(comp_word)
    round1_remaining_word_scores['word'].append(word)
    round1_remaining_word_scores['remaining_word_count'].append(len(remaining_words))
    round1_remaining_word_scores['remaining_words'].append(remaining_words)
    all_letters = list(word)
    col_counter = 1
    for letter in all_letters:
        col_name = 'lett_' + str(col_counter)
        round1_remaining_word_scores[col_name].append(letter_lookup[letter])
        
        pos_proba = get_position_probability(letter,col_counter - 1,five_letter_words)
        pos_col_name = 'pos_' + str(col_counter)
        round1_remaining_word_scores[pos_col_name].append(pos_proba)
        col_counter += 1
    counter += 1
    if counter % 1000 == 0:
        print('Records finished: {}'.format(counter))                
```

```python
scored_df = pd.DataFrame(round1_remaining_word_scores).sort_values(by='remaining_word_count', ascending=True).reset_index()
scored_df.loc[5:,].head(50)
```

```python
outcomes = [0,1,2] #0, gray | 1, yellow | 2, green
combinations = [combo for combo in itertools.product(outcomes, repeat=5)]
combinations

def generate_match_matrix(start_word, word_list):
    outcome_distributions = {
    }
    start_word = start_word.lower()
    start_word_chars = list(start_word)
    match_words = {}
    for combo in combinations:
        match_words[combo] = {
            'count' : 0,
            'words' : []
        }
    
    for word in word_list:
        ref_word_chars = list(word.lower())
        for combo in combinations:
            is_remaining_word = 0
            combo_pos_counter = 0
            match_score = 0
            for outcome in combo:
                if outcome == 1:
                    if ref_word_chars[combo_pos_counter] in start_word_chars and ref_word_chars[combo_pos_counter] != start_word_chars[combo_pos_counter]:
                        match_score += 1
                elif outcome == 2:
                    if ref_word_chars[combo_pos_counter] == start_word_chars[combo_pos_counter]:
                        match_score += 1
                elif outcome == 0:
                    if ref_word_chars[combo_pos_counter] not in start_word_chars:
                        match_score += 1
                combo_pos_counter += 1
                    
            if match_score == 5:
                is_remaining_word = 1
                match_words[combo]['count'] += 1
                match_words[combo]['words'].append(word)
    return match_words
            

all_word_distributions = {}
counter = 0
for word in five_letter_words:
    all_word_distributions[word] = generate_match_matrix(word, five_letter_words)     
    counter += 1
    if counter % 10 == 0:
        print('{} Words Finished!'.format(counter))
```

```python
df_dict = {
    'word' : [],
    'outcome' : [],
    'remaining_word_count' : []
}

for key, value in all_word_distributions.items():
    for outcome in value.keys():
        df_dict['word'].append(key)
        df_dict['outcome'].append(outcome)
        df_dict['remaining_word_count'].append(value[outcome]['count'])
        
        
```
